```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generar variables universales para la temperatura (ejemplo de 0 a 100)
# Este es el rango de valores sobre el que se definirán las funciones de pertenencia.
x_universe = np.linspace(0, 100, 1000)

# Definir tres (3) funciones difusas para representar los tres estados "bajo", "medio" y "alto".
# Se utiliza la función trimf (función de pertenencia triangular)
# Cada función se define por tres puntos: el inicio del triángulo, el pico y el final.

# Función para el estado "bajo": va de 0 a 50, con el pico en 0.
# Esto significa que la pertenencia es 1 en 0 y disminuye linealmente hasta 0 en 50.
bajo = fuzz.trimf(x_universe, [0, 0, 50])

# Función para el estado "medio": va de 0 a 100, con el pico en 50.
# La pertenencia es 0 en 0, aumenta a 1 en 50, y disminuye a 0 en 100.
medio = fuzz.trimf(x_universe, [0, 50, 100])

# Función para el estado "alto": va de 50 a 100, con el pico en 100.
# La pertenencia es 0 en 50 y aumenta linealmente hasta 1 en 100.
alto = fuzz.trimf(x_universe, [50, 100, 100])

# Graficar las funciones difusas en un solo eje.
# Esto visualiza cómo cada estado se asigna a diferentes valores en el universo de discurso,
# permitiendo una representación flexible y subjetiva de los valores.
fig, ax = plt.subplots(figsize=(8, 5)) # Crea una figura y un conjunto de subplots

ax.plot(x_universe, bajo, 'b', linewidth=1.5, label='Bajo') # Grafica la función "bajo" en azul
ax.plot(x_universe, medio, 'g', linewidth=1.5, label='Medio') # Grafica la función "medio" en verde
ax.plot(x_universe, alto, 'r', linewidth=1.5, label='Alto')   # Grafica la función "alto" en rojo

ax.set_title('Funciones de pertenencia') # Establece el título del gráfico
ax.legend() # Muestra la leyenda de las etiquetas
ax.set_xlabel('Valor en el universo de discurso') # Etiqueta para el eje X
ax.set_ylabel('Grado de pertenencia') # Etiqueta para el eje Y
ax.grid(True) # Añade una cuadrícula al gráfico
plt.show() # Muestra el gráfico
```
